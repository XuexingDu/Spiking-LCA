```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
colors = ['#8ECFC9',  "#FFA500", '#FA7F6F', '#82B0D2','#BEB8DC','#E7DAD2']
# colors   = ['#2878b5', '#9ac9db', '#f8ac8c', '#c82423','#ff8884']
markers = [(5,1), 'o', 's', 'D','*']

plt_style = {
    'figure.autolayout' : True,
    'font.size' : 18,
    'lines.linewidth' : 3,
    'xtick.labelsize' : 'medium',
    'ytick.labelsize' : 'medium',
    'legend.fontsize' : 'small',
    # 'axes.spines.top' : False,
    # 'axes.spines.right' : False,
    'xtick.labelsize' : 24,
    'ytick.labelsize' : 24,
    'xtick.major.size' : 6,
    'ytick.major.size' : 6,
    'legend.fontsize' : 18,
    'axes.labelsize' : 24,
    'axes.titlesize' : 24,
    'lines.markersize' : 12,
    # 'font.family': "Times New Roman"
    }
for key, val in plt_style.items():
    plt.rcParams[key] = val
```

```python
loaded_data = np.load('Figure/Fig5a.npz')
lasso_time = loaded_data['lasso_time']
nmse_lasso = loaded_data['nmse_lasso']
exp_time   = loaded_data['exp_time']
nmse_exp   = loaded_data['nmse_exp']
log_time   = loaded_data['log_time']
nmse_log  = loaded_data['nmse_log']
atan_time = loaded_data['atan_time']
nmse_atan = loaded_data['nmse_atan']
```

```python
plt.figure(figsize=(9,6))
plt.plot(lasso_time,nmse_lasso, color=colors[0], marker=markers[0], markerfacecolor='none', label='SLCA-LASSO',markersize = 15)
plt.plot(exp_time,nmse_exp,   color=colors[1], marker=markers[1], markerfacecolor='none', label='SLCA-Exp')
plt.plot(log_time,nmse_log,   color=colors[2], marker=markers[2], markerfacecolor='none', label='SLCA-Log')
plt.plot(atan_time,nmse_atan,  color=colors[3], marker=markers[3], markerfacecolor='none', label='SLCA-Atan')
plt.legend()
plt.xlim([0,2])
plt.ylim([-28,1])
plt.yticks([0, -10,-20])
plt.xticks([0,1,2])
plt.xlabel('Execution Time (s)')
plt.ylabel('NMSE (dB)')
plt.savefig('Figure/multi_task.pdf', format='pdf')
plt.savefig('Figure/multi_task.eps', format='eps')
plt.savefig('Figure/multi_task.png', format='png')

print(nmse_lasso[-1],nmse_exp[-1],nmse_log[-1],nmse_atan[-1])
```

```python
loaded_data = np.load('Figure/Fig5b.npz')
NMSE_matrix = loaded_data['NMSE_matrix']
sparsity    = loaded_data['sparsity']
```

```python
x_ticks = sparsity * 100
x_tick_labels = ['{:.0f}%'.format(x) for x in x_ticks]
# Select every other label for x-ticks
plt.figure(figsize=(9,6))
plt.plot(sparsity, NMSE_matrix[0,:], color=colors[0], marker=markers[0], markerfacecolor='none', label='SLCA-LASSO',markersize = 15)
plt.plot(sparsity, NMSE_matrix[1,:], color=colors[1], marker=markers[1], markerfacecolor='none', label='SLCA-Exp')
plt.plot(sparsity, NMSE_matrix[2,:], color=colors[2], marker=markers[2], markerfacecolor='none', label='SLCA-Log')
plt.plot(sparsity, NMSE_matrix[3,:], color=colors[3], marker=markers[3], markerfacecolor='none', label='SLCA-Atan')
plt.legend()
plt.xticks(sparsity[1::2], x_tick_labels[1::2])
# plt.xticks(sparsity[1::2], x_tick_labels[1::2])
plt.ylim([-28,1])
# plt.ylim([-25,-10])
plt.yticks([0, -10,-20])
plt.xlabel(r'Sparsity ($\%$)')
# plt.xlabel(r'Sparsity ($\%$)')
plt.ylabel('NMSE (dB)')
plt.savefig('Figure/sparsity.pdf', format='pdf')
plt.savefig('Figure/sparsity.eps', format='eps')
plt.savefig('Figure/sparsity.png', format='png')
```
